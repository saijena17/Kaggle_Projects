```python
data=pd.read_csv('../input/train.csv')
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
X=X.loc[:, X.columns != 'ID']  #eliminating the Id column
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
X=pd.DataFrame(X)
#performing PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=50)
X=pca.fit_transform(X)
explnVR=pca.explained_variance_ratio_
plt.bar(range(0,50),explnVR)
plt.show()
```

```python
#looking at the final X dataset
X=pd.DataFrame(X)
X.head(3)
```

```python
#splitting of dataset
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python


                                    #LOGISTIC REGRESSION

```

```python
from sklearn.linear_model import LogisticRegression
cls_lr=LogisticRegression(random_state=0)
cls_lr.fit(X_train,y_train)
```

```python
#prediction
y_pred=cls_lr.predict(X_test)
#confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
#Let us have a look at the y set
y_train
```

```python
'''looking at the no of 0 we have to compare the no. of accuarte 1 we get from our model
 So lets build a code to do that'''
```
