```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing dataset
df=pd.read_csv("../input/train.csv")
```

```python
#knowning dataset
df.isnull().sum()
```

```python
df.nunique()
```

```python
df.columns
```

```python
#Defining X and y
X=df[['bone_length', 'rotting_flesh', 'hair_length', 'has_soul',
       'color']]
y=df[['type']]
```

```python
#applying labelencoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X[['color']]=le.fit_transform(X[['color']])
```

```python
X.head(3)
```

```python
#applying OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(categorical_features=[4])
X=ohe.fit_transform(X).toarray()
```

```python
#removing the Dummy variable
a=pd.DataFrame(X)
a=a[[1,2,3,4,5,6,7,8,9]]
a.head(3)
```

```python
#Applying Label Encoder to encode the y set
LE=LabelEncoder()
y['type']=LE.fit_transform(y['type']) 
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
Xs=sc_X.fit_transform(X)
pdf=pd.DataFrame(Xs)
pds=pd.DataFrame(X)
```

```python
#splitting of set
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(pdf,y,test_size=0.2)
```

```python
                                        
    
                                        #Classification Models

        
```

```python
                                    
                                                #LOGISTIC REGRESSION
        
```

```python
#fitting logistic regression to training set
from sklearn.linear_model import LogisticRegression
cls_lr=LogisticRegression()
cls_lr.fit(X_train,y_train)
```

```python
#predicting the result
y_pred=cls_lr.predict(X_test)
```

```python
#making Confusion Matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python

                                                #k-NNeighbors
    
```

```python
from sklearn.neighbors import KNeighborsClassifier
cls_knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
cls_knn.fit(X_train,y_train)
```

```python
#predicting the result
y_pred_knn=cls_knn.predict(X_test)
```

```python
#making Confusion Matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred_knn)
cm
```

```python

                                                            #Decision Tree
```

```python
from sklearn.tree import DecisionTreeClassifier
cls_dt=DecisionTreeClassifier(criterion='entropy',random_state=0)
cls_dt.fit(X_train,y_train)
```

```python
#prediction
y_pred_dt=cls_dt.predict(X_test)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred_dt)
cm
```

```python

                                        #RandomForest
```

```python
from sklearn.ensemble import RandomForestClassifier
cls_rf=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
cls_rf.fit(X_train,y_train)
```

```python
#prediction
y_pred_rf=cls_rf.predict(X_test)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred_rf)
cm
```

```python

                                        #NAIVE BAYES
```

```python
from sklearn.naive_bayes import GaussianNB
cls_nb=GaussianNB()
cls_nb.fit(X_train,y_train)
```

```python
#prediction
y_pred_nb=cls_nb.predict(X_test)
#confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred_nb)
cm
```

```python

                                            #SVM
```

```python
from sklearn.svm import SVC
cls_svm=SVC(kernel='linear',random_state=0)
cls_svm.fit(X_train,y_train)
```

```python
#prediction
y_pred_svm=cls_svm.predict(X_test)
#confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred_svm)
cm
```

```python

                                #K-FOLD CROSS VALIDATION
```

```python
#Applying K-fold cross validation
from sklearn.model_selection import cross_val_score
accuracies_lr=cross_val_score(estimator=cls_lr,X=X_train,y=y_train,cv=10)
```

```python
accuracies_lr.mean()
```

```python
#Applying Grid-Search
from sklearn.model_selection import GridSearchCV
param=[{'C':[1,10,100,1000],'kernel':['linear'],'gamma':[0.5,0.1,0.01,0.001]}]
grid_seach=GridSearchCV(estimator=cls_svm,param_grid=param,scoring='accuracy',cv=10)
```

```python
grid_seach=grid_seach.fit(X_train,y_train)
best_accuracy=grid_seach.best_score_
best_parameters=grid_seach.best_params_
```
