```python
#importing libraries
import nltk
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing Dataset
data=pd.read_csv('../input/spooky-author/train.csv')
X=data['text']
y=data['author']
```

```python
len(data)
```

```python
#importing some NLP libraries
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
import re
from nltk.corpus import stopwords
```

```python
#DataPreprocessing

#Cleaning the texts
corpus=[]
for i in range(0,19579):
    review=re.sub('[^a-zA-Z]',' ',X[i])
    review=review.lower()
    review=review.split()
    review=[words for words in review if not words in stopwords.words('english')]
    review=[ps.stem(words) for words in review]
    review=' '.join(review)
    corpus.append(review)
    
```

```python
'''Let us have a look at the corpus'''
```

```python
corpus
```

```python
#creating bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1500)
X=cv.fit_transform(corpus).toarray()
```

```python
x=pd.DataFrame(X,columns=cv.get_feature_names())
x.head(3)
```

```python
#applying Label Encoder to y
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
#splitting X and y to train and test set
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
    
    
                                           #CLASSIFIERS
    
    
```

```python

                                        #LOGISTIC REGRESSION
```

```python
#fitting the model
from sklearn.linear_model import LogisticRegression
cls_lr=LogisticRegression(random_state=0)
cls_lr.fit(X_train,y_train)
```

```python
#prediction and confusion matrix 
y_pred=cls_lr.predict(X_test)

from sklearn.metrics import confusion_matrix
cm_lr=confusion_matrix(y_test,y_pred)
cm_lr   #74.8 accuracy
```

```python

                                                            #KNNeighbors
```

```python
from sklearn.neighbors import KNeighborsClassifier
cls_knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
cls_knn.fit(X_train,y_train)
```

```python
#prediction and confusion matrix 
y_pred=cls_knn.predict(X_test)

from sklearn.metrics import confusion_matrix
cm_knn=confusion_matrix(y_test,y_pred)
cm_knn    #45%
```

```python

                                                #DECISION TREE
```

```python
from sklearn.tree import DecisionTreeClassifier
cls_dt=DecisionTreeClassifier(criterion='entropy',random_state=0)
cls_dt.fit(X_train,y_train)
```

```python
#prediction and confusion matrix 
y_pred=cls_dt.predict(X_test)

#confusion matrix
from sklearn.metrics import confusion_matrix
cm_dt=confusion_matrix(y_test,y_pred)
cm_dt    #58%
```

```python

                                        #RANDOM FOREST CLASSIFIER
```

```python
from sklearn.ensemble import RandomForestClassifier
cls_rf=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
cls_rf.fit(X_train,y_train)
```

```python
#prediction and confusion matrix 
y_pred=cls_rf.predict(X_test)

#confusion matrix
from sklearn.metrics import confusion_matrix
cm_rf=confusion_matrix(y_test,y_pred)
cm_rf    #64%
```

```python

                                        #NAIVE BAYES
```

```python
from sklearn.naive_bayes import GaussianNB
cls_nb=GaussianNB()
cls_nb.fit(X_train,y_train)
```

```python
#prediction and confusion matrix 
y_pred=cls_nb.predict(X_test)

#confusion matrix
from sklearn.metrics import confusion_matrix
cm_nb=confusion_matrix(y_test,y_pred)
cm_nb    #69%
```

```python

                                    #SVM-KERNEL
```

```python
from sklearn.svm import SVC
cls_svm=SVC(kernel='linear',random_state=0)
cls_svm.fit(X_train,y_train)
```

```python
#prediction and confusion matrix 
y_pred=cls_svm.predict(X_test)

#confusion matrix
from sklearn.metrics import confusion_matrix
cm_svm=confusion_matrix(y_test,y_pred)
cm_svm    #72.77%
```
