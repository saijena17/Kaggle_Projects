```python
X=train[['Elevation', 'Aspect', 'Slope',
       'Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology',
       'Horizontal_Distance_To_Roadways', 'Hillshade_9am', 'Hillshade_Noon',
       'Hillshade_3pm', 'Horizontal_Distance_To_Fire_Points',
       'Wilderness_Area1', 'Wilderness_Area2', 'Wilderness_Area3',
       'Wilderness_Area4', 'Soil_Type1', 'Soil_Type2', 'Soil_Type3',
       'Soil_Type4', 'Soil_Type5', 'Soil_Type6', 'Soil_Type7', 'Soil_Type8',
       'Soil_Type9', 'Soil_Type10', 'Soil_Type11', 'Soil_Type12',
       'Soil_Type13', 'Soil_Type14', 'Soil_Type15', 'Soil_Type16',
       'Soil_Type17', 'Soil_Type18', 'Soil_Type19', 'Soil_Type20',
       'Soil_Type21', 'Soil_Type22', 'Soil_Type23', 'Soil_Type24',
       'Soil_Type25', 'Soil_Type26', 'Soil_Type27', 'Soil_Type28',
       'Soil_Type29', 'Soil_Type30', 'Soil_Type31', 'Soil_Type32',
       'Soil_Type33', 'Soil_Type34', 'Soil_Type35', 'Soil_Type36',
       'Soil_Type37', 'Soil_Type38', 'Soil_Type39', 'Soil_Type40']]
y=train['Cover_Type']
```

```python
#Making the Cover type starting from 0 and not 1
for i in range(0,len(y)):
    y[i]=y[i]-1
```

```python
#Feature Scaling on X
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
X=pd.DataFrame(X)
```

```python


                                                  #ARTIFICIAL NEURAL NETWORK

```

```python
#splitting of dataset
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
y_train=pd.DataFrame(y_train)
```

```python
#importing deep learning libraries
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#initialization of ANN
cls=Sequential()
```

```python
#adding the first layer
cls.add(Dense(output_dim=32,input_dim=54,activation='relu',init='uniform'))
```

```python
#adding second Layer
cls.add(Dense(output_dim=32,activation='relu',init='uniform'))
```

```python
#adding third Layer
cls.add(Dense(output_dim=32,activation='relu',init='uniform'))
```

```python
#adding final Layer
cls.add(Dense(output_dim=7,activation='softmax',init='uniform'))
```

```python
#compiling ANN
cls.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#fitting ANN to the training set
cls.fit(X_train,y_train,batch_size=8,nb_epoch=200)
```

```python
''' we got a predicted accuracy of 87.9% '''
```

```python
#predicting the result
y_pred=cls.predict(X_test)
```

```python
#converting the result of ANN to categorical numbers
y_predvalue=[]
for i in range(0,len(y_pred)):
    for j in range(0,7):
        if y_pred[i][j]==y_pred[i].max():
            y_predvalue.append(j)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predvalue)
cm
```

```python

                                                        #APPLYING K-FOLD
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
#defining classifier function
def build_classifier():
    cls=Sequential()
    cls.add(Dense(output_dim=32,input_dim=54,activation='relu',init='uniform'))
    cls.add(Dense(output_dim=32,activation='relu',init='uniform'))
    cls.add(Dense(output_dim=32,activation='relu',init='uniform'))
    cls.add(Dense(output_dim=7,activation='softmax',init='uniform'))
    cls.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
    return cls
```

```python
#wrapping whole thing
clss=KerasClassifier(build_fn=build_classifier,batch_size=8,nb_epoch=200)
accuracies=cross_val_score(estimator=clss,X=X_train,y=y_train,cv=10)
```

```python
accuracies.mean()
```

```python
accuracies.std()
```

```python
                                        #parameter tunning of ANN
```

```python
#rebulding the function
def build_classifier(optimizer):
    cls=Sequential()
    cls.add(Dense(output_dim=32,input_dim=54,activation='relu',init='uniform'))
    cls.add(Dense(output_dim=32,activation='relu',init='uniform'))
    cls.add(Dense(output_dim=32,activation='relu',init='uniform'))
    cls.add(Dense(output_dim=7,activation='softmax',init='uniform'))
    cls.compile(optimizer=optimizer,loss='sparse_categorical_crossentropy',metrics=['accuracy'])
    return cls
```

```python
#tunning
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
classifier=KerasClassifier(build_fn=build_classifier)
param={'batch_size':[25,32],'nb_epoch':[100,500],'optimizer':['adam','rmsprop']}
grid_search=GridSearchCV(estimator=classifier,param_grid=param,scoring='accuracy',cv=10)
grid_search=grid_search.fit(X_train,y_train)
```

```python
best_param=grid_search.best_params_
best_param
```

```python
best_accuracy=grid_search.best_score_
best_accuracy
```

```python
''' By applying ANN we got an accuracy of 63%
    Now let us see the prediction done by Machine learning algorithm '''
```
